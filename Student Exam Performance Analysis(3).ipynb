<h2>Data Cleaning</h2>

<h4>Handling Missing Values</h4>

```python
data = data.fillna('NA')
data
```

<h4>Categorical Encoding</h4>

```python
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})
data['Pass or Fail'] = data['Pass or Fail'].map({'Pass': 1, 'Fail': 0})
data['Receiving Tuition'] = data['Receiving Tuition'].map({'Yes': 1, 'No': 0})
data['Enrolled in Test Preparation Course'] = data['Enrolled in Test Preparation Course'].map({'Yes': 1, 'No': 0})
```

<h4>Removing Duplicate Rows</h4>

```python
data = data.drop_duplicates()
```

<h4>Dropping Irrelevant Columns</h4>

```python
data = data.drop(['Roll No'], axis=1)
```

<h4>Renaming Long Column Names</h4>

```python
data.rename(columns={
    'Hours of Study per Day': 'Study Hours',
    'Hours of Sleep per Day': 'Sleep Hours',
    'Hours of Self-Study per Day':'Self Study',
    'Assignment Submission Rate (%)': 'Assignment Rate',
    'Attendance (%)': 'Attendance'
}, inplace=True)
data
```

<h2>Data Filtering</h2>

```python
#Students with Actual Percentage above 85
high_scorers = data[data['Actual Percentage'] > 85][['Name', 'Actual Percentage']]
print("High Scorers (Name & Actual Percentage > 85%):\n", high_scorers.head())
```

```python
#Female Students Who Passed and Study More Than 4 Hours

female_pass = data[(data['Gender'] == 0) & (data['Pass or Fail'] == 1) & (data['Study Hours'] > 4)]

# Selecting academic columns only
academic_columns = [
    'Study Hours', 'Sleep Hours', 'Attendance',
    'Assignment Rate', 'Actual Percentage', 'Expected Percentage',
    'Pass or Fail', 'Enrolled in Test Preparation Course',
    'Receiving Tuition', 'Previous Exam Performance'
]

female_academic = female_pass[academic_columns]
print("Academic details of female students who passed and study > 4 hours:\n", female_academic.tail())
```

```python
#Students Enrolled in Test Prep and Receiving Tuition

prep_and_tuition = data[
    (data['Enrolled in Test Preparation Course'] == 1) & 
    (data['Receiving Tuition'] == 1)
][['Name', 'Actual Percentage']]

print("Students enrolled in test prep and receiving tuition:\n", prep_and_tuition)
```

<H2>Data Visualizations</H2>

<h3><i>1. Scatter Plot: Attendance vs Actual and Expected Percentage</h3>

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Attendance', y='Actual Percentage', data=data, label='Actual', color='blue')
sns.scatterplot(x='Attendance', y='Expected Percentage', data=data, label='Expected', color='green')
plt.xlabel('Attendance')
plt.ylabel('Percentage')
plt.title('Attendance vs Actual and Expected Percentage')
plt.legend()
plt.show()

```

<h3><i>2. Scatter Plot: Hours of Study vs Actual and Expected Percentage</h3>

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Study Hours', y='Actual Percentage', data=data, color='blue', label='Actual')
sns.scatterplot(x='Study Hours', y='Expected Percentage', data=data, color='green', label='Expected')
plt.title('Study Hours vs Actual and Expected Percentage')
plt.legend()
plt.show()
```

<H3><i>3. Bar Plot: Previous Exam Performance vs Actual Percentage</H3>

```python
# Round previous performance for grouping
data['Previous Performance Group'] = data['Previous Exam Performance'].round(-1)

# Group by Previous Performance Group
grouped = data.groupby('Previous Performance Group')[['Actual Percentage', 'Expected Percentage']].mean().reset_index()

# Melt for side-by-side bar plot
melted = grouped.melt(id_vars='Previous Performance Group', 
                      value_vars=['Actual Percentage', 'Expected Percentage'],
                      var_name='Type', value_name='Percentage')

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Previous Performance Group', y='Percentage', hue='Type', data=melted)
plt.xlabel('Previous Exam Performance Group')
plt.ylabel('Percentage')
plt.title('Previous Exam Performance vs Actual and Expected Percentage')
plt.legend(title='Score Type')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

<h3><i>4. Box Plot: Class Participation vs Actual Percentage</h3>

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class Participation', y='Actual Percentage', data=data)
plt.title('Class Participation vs Actual Percentage')
plt.xticks(rotation=45)
plt.show()
```

<h3><i>5. Violin Plot: Stress Level vs Actual Percentage</h3>

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Stress Level', y='Actual Percentage', data=data)
plt.title('Stress Level vs Actual Percentage')
plt.show()
```

<h3><i>6. Line Plot: Assignment Submission vs Actual and Expected Percentage</h3>

```python
import matplotlib.pyplot as plt

# Round assignment submission rate to nearest 10
data['Assignment Group'] = data['Assignment Rate'].round(-1)

# Group data and calculate average actual & expected percentages
assignment_grouped = data.groupby('Assignment Group')[['Actual Percentage', 'Expected Percentage']].mean()

# Plot
plt.figure(figsize=(10,6))
plt.plot(assignment_grouped.index, assignment_grouped['Actual Percentage'], marker='o', label='Actual', color='blue')
plt.plot(assignment_grouped.index, assignment_grouped['Expected Percentage'], marker='o', label='Expected', color='green')
plt.xlabel('Assignment Submission Rate Group (%)')
plt.ylabel('Average Percentage')
plt.title('Assignment Submission vs Actual and Expected Percentage (Grouped)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
